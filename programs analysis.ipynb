## Includes all program donations and other donations. Not all donations are linked to a program.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
import pandas as pd
pd.set_option('display.max_rows', None)
df = pd.read_csv('TGE programs analysis.csv',)
df
```

```python
df = df.rename(columns={'Unnamed: 0':'Donation Date','Unnamed: 1':'Donation Amount','Unnamed: 2':'Program'})
```

## I renamed the columns and got rid of the 1st row since those are the names of the data points I pulled from fundly

```python
df = df[1:]
```

```python
df['Donation Amount'] = df['Donation Amount'].astype(float)
```

## There are 993 records in the data, most donations fell within 50 - 450 but a donation around 100 seems to be most common, and 41,000 was the largest donation

```python
df['Donation Amount'].describe()
```

## This plot contains the same information as the data above

```python
sns.boxplot(df['Donation Amount'])
```

```python
outliers = df[df['Donation Amount'] > 450]
```

## The next largest donations above 450 is 475. 749 donations are between 50 - 450 and 244 donations are between 475 and 2241. Outliers isn't really the right word, it's more major donors but even this could be filtered further depending on how a major donor is defined.

```python
outliers['Donation Amount'].describe()
```

```python
sns.boxplot(outliers['Donation Amount'])
```

```python
# 16
df['Program'].value_counts()
```

```python
# 13
outliers['Program'].value_counts()
```

```python
outliers.groupby('Program')['Donation Amount'].sum().plot(kind='bar')
plt.ylabel('Donation Amount in $')
plt.show()
outliers.groupby('Program')['Donation Amount'].count().plot(kind='bar')
plt.ylabel('# of unique donations within the filtered dataset')
plt.show()
```

## By comparison the first 3 programs made the most money while having a lower amount of unique donations. The next program that seems to have made more while having a lower number of unique donations is the documentary

```python
df.groupby('Program')['Donation Amount'].sum().plot(kind='bar')
plt.ylabel('Donation Amount in $')
plt.show()
df.groupby('Program')['Donation Amount'].count().plot(kind='bar')
plt.ylabel('# of unique donations within the whole dataset')
plt.show()
```

```python
df['Donation Date'] = pd.to_datetime(df['Donation Date'])
```

## 123 donations made on 2020-12-01.

## 499 dates are different and 494 dates are the same. This includes the 123 dates of 12/01/2021.  This means that 494 dates had more than 1 donation not that 494 donations fell on one dates.

```python
df['Donation Date'].value_counts()
```

```python
df['Donation Date'].describe()
```

```python
df['year'] = df['Donation Date'].dt.year
df['month'] = df['Donation Date'].dt.month
df['day'] = df['Donation Date'].dt.day
```

```python
df.groupby( 'month')['Donation Amount'].sum().plot(kind='bar')
plt.ylabel('Donation Amount in $ for each month')
plt.show()
df.groupby( 'month')['Donation Amount'].count().plot(kind='bar')
plt.ylabel('# of unique donations within the whole dataset/month')
plt.show()
```

```python
df.groupby( 'day')['Donation Amount'].sum().plot(kind='bar')
plt.show()
df.groupby( 'day')['Donation Amount'].count().plot(kind='bar')
plt.show()
```

```python
df.groupby( 'year')['Donation Amount'].sum().plot(kind='bar')
plt.ylabel('Donation Amount in $ for each year')
plt.show()
df.groupby( 'year')['Donation Amount'].count().plot(kind='bar')
plt.ylabel('# of unique donations within the whole dataset/year')

plt.show()
```

```python
## Some of the donations overlap but this graph shows the amount of donations(x axis) per year(y axis) per program 
```

```python
plt.figure(figsize=(15,12))
sns.scatterplot(x='Donation Amount', y='year', hue='Program', size="Program",
            sizes=(50, 500), alpha=.8,palette="muted", data=df)
plt.legend(bbox_to_anchor=(1.01, 1),
           borderaxespad=0)
```
